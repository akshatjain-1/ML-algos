```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/datasets/Salary_Data.csv')
dataset
```

```python
X = dataset.iloc[:,:-1].values
```

```python
y = dataset.iloc[:,1].values
```

```python
X
```

```python
y
```

```python
#splitting into training and testing

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=101 )
```

```python
X_train
```

```python
X_test
```

```python
y_train.shape
```

```python
y_test
```

```python
#TRAINNG THE SIMPLE LINEAR MODEL ON TRANIG SET
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()   #Instantitaing the model / model class
```

```python
reg.fit(X_train, y_train)
```

```python
#Predict the test results
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
y_test
```

```python
#Visualizing the training set results
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.scatter(X_test, y_pred, color = 'green')
plt.plot(X_train, reg.predict(X_train), color='yellow')
plt.title('savalry vs expectation')
plt.xlabel('yoe')
plt.ylabel('salary')
plt.show()
```
