```python
df = pd.read_csv('/content/drive/MyDrive/datasets/Advertising.csv')
df
```

```python
df.info()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize= (16,6))

axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].set_title('TV VS Sales')

axes[1].plot(df['radio'], df['sales'], 'o', color = 'green')
axes[1].set_title('radio VS sales')

axes[2].plot(df['newspaper'], df['sales'], 'o', color= 'red')
axes[2].set_title('newspaper VS sales')


axes[3].plot(df['TV'], df['sales'], 'o')

axes[3].plot(df['radio'], df['sales'], 'o', color = 'green')

axes[3].plot(df['newspaper'], df['sales'], 'o', color= 'red')
axes[3].set_title('All spend VS sales')
```

```python
X = df.drop('sales', axis =1)
y = df['sales']
```

```python
X
```

```python
y
```

```python
#Train and test splits
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Create a model
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```
