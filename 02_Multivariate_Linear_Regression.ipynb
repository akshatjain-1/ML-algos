```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/Advertising.csv')
df
```

```python
df.info()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize= (16,6))

axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].set_title('TV VS Sales')

axes[1].plot(df['radio'], df['sales'], 'o', color = 'green')
axes[1].set_title('radio VS sales')

axes[2].plot(df['newspaper'], df['sales'], 'o', color= 'red')
axes[2].set_title('newspaper VS sales')


axes[3].plot(df['TV'], df['sales'], 'o')

axes[3].plot(df['radio'], df['sales'], 'o', color = 'green')

axes[3].plot(df['newspaper'], df['sales'], 'o', color= 'red')
axes[3].set_title('All spend VS sales')
```

```python
X = df.drop('sales', axis =1)
y = df['sales']
```

```python
X
```

```python
y
```

```python
#Train and test splits
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Create a model
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#Understanding and utilizing Model

###Performance analysis
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
```

```python
MAE
```

```python
avg_sales = df['sales'].mean()
```

```python
percent_error = (MAE/avg_sales)*100
percent_error
```

```python
MSE
```

```python
RMSE
```

#Plotting residuals

```python
test_res = y_test - y_pred
```

```python
sns.scatterplot(x = y_test, y = test_res)
plt.axhline(y=0, color = 'red', ls = '--')
```

```python
sns.kdeplot(test_res)
```

#Retraining the model on the Full Data

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)
```

```python
y_hat = final_model.predict(X)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize= (16,6))

axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].plot(df['TV'], y_hat, 'o')
axes[0].set_title('TV VS Sales')

axes[1].plot(df['radio'], df['sales'], 'o', color = 'green')
axes[1].plot(df['radio'], y_hat, 'o', color = 'yellow')
axes[1].set_title('radio VS sales')

axes[2].plot(df['newspaper'], df['sales'], 'o', color= 'red')
axes[2].plot(df['newspaper'], y_hat, 'o', color= 'blue')

axes[2].set_title('newspaper VS sales')


axes[3].plot(df['TV'], df['sales'], 'o')
axes[3].plot(df['radio'], df['sales'], 'o', color = 'green')
axes[3].plot(df['newspaper'], df['sales'], 'o', color= 'red')
axes[3].set_title('All spend VS sales')
```

```python
residuals = y_hat - y
sns.scatterplot(x=y, y=residuals)
```
